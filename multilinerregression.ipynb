```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Experience': [1, 2, 3, 5, 7],
    'Education_Level': [1, 1, 2, 2, 3],
    'Projects': [2, 3, 4, 6, 8],
    'Salary': [25, 30, 40, 55, 75]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['Experience', 'Education_Level', 'Projects']]
y = df['Salary']
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python

# Predict salary for:
# Experience = 4 years, Education Level = 2, Projects = 5
prediction = model.predict([[4, 2, 5]])
print("Predicted Salary (in Lakhs):", prediction)

```

```python

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Use only 2 features
X = df[['Experience', 'Projects']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

# Create grid
x1, x2 = np.meshgrid(
    np.linspace(X['Experience'].min(), X['Experience'].max(), 10),
    np.linspace(X['Projects'].min(), X['Projects'].max(), 10)
)

# Predict plane
y_pred = model.predict(np.c_[x1.ravel(), x2.ravel()]).reshape(x1.shape)

# Plot
ax = plt.figure().add_subplot(projection='3d')
ax.scatter(X['Experience'], X['Projects'], y)
ax.plot_surface(x1, x2, y_pred, alpha=0.5)

plt.show()
```
