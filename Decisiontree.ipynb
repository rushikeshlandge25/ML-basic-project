```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python

data = {
    'Hours': [2,3,4,5,6,7,8,9],
    'Attendance': [60,62,65,70,75,80,85,90],
    'Prev_Score': [45,50,55,60,65,70,80,85],
    'Result': [0,0,0,1,1,1,1,1]
}
df=pd.DataFrame(data)
df
```

```python
model=DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(df[['Hours','Attendance','Prev_Score']],df['Result'])
```

```python
prediction=model.predict([[10,80,90]])
if prediction==0:
  print('Fail')
else:
  print('Pass')
```

```python
plt.figure(figsize=(14,8))
plot_tree(
    model,
    feature_names=['Hours','Attendance','Prev_Score'],
    class_names=['Fail','Pass'],
    filled=True
)
plt.show()
```
