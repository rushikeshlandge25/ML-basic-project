```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le=LabelEncoder()
df['time']=le.fit_transform(df['time'])
df
```

```python
x=df.drop('time',axis=1)
y=df['time']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder ,StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
categotrical_features=['sex','smoker','day']
numerical_features=['total_bill','tip','size']
```

```python

df[categotrical_features]
```

```python
pipe_num=Pipeline([
    ('simpleimputer',SimpleImputer(strategy='median')),
    ('scalar',StandardScaler())
])
```

```python
pipe_cat=Pipeline([
    ('simpleimputer',SimpleImputer(strategy='most_frequent')),
    ('onehotencoder',OneHotEncoder())
])
```

```python
col_trans=ColumnTransformer([
    ('pipe_num',pipe_num,numerical_features),
    ('pipe_cat',pipe_cat,categotrical_features)
])
```

```python
col_trans.fit_transform(x_train,y)
```

```python
col_trans
```

```python
x_train=col_trans.fit_transform(x_train)
```

```python
x_test=col_trans.transform(x_test)
```

```python
x_train
```

```python
models={
    'RandomForestClassifier':RandomForestClassifier(),
    'SupportVectorClassifier':SVC(),
    'DecisionTreeClassifier':DecisionTreeClassifier()


}
print(list(models.keys())[0])
print(list(models.keys())[1])
print(list(models.keys())[2])
```

```python
def evaluate_model(x_train,x_test,y_train,y_test,models):
  report={}
  for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    acc_score=accuracy_score(y_test,y_pred)*100
    report[list(models.keys())[i]]=acc_score
  return report
```

```python
evaluate_model(x_train,x_test,y_train,y_test,models)
```
